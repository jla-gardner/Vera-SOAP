```python
from sphere_sampling import cart2sphr, sphr2cart, \
        sample_unit_sphere_random, sample_unit_sphere_uniform
import plotly.graph_objects as go
import numpy as np


def plot_points(x, y, z):
    fig = go.Figure(data=[go.Scatter3d(
        x=x, y=y, z=z, mode='markers',
        marker=dict(size=4, opacity=0.8)
    )])
    fig.update_layout(
        autosize=False,
        width=350,
        height=350,
        margin=dict(l=0, r=0, b=0, t=0)
    )
    fig.show()
```

```python
cart, sphr = sample_unit_sphere_random(N=400)

x, y, z = cart.T
plot_points(x, y, z)
```

```python
cart, sphr = sample_unit_sphere_uniform(N=400)

x, y, z = cart.T
plot_points(x, y, z)
```

```python
points_cart = np.random.randn(20, 3)
points_sphr = cart2sphr(points_cart)
points_cart_ = sphr2cart(points_sphr)

x, y, z = points_cart.T
xx, yy, zz = points_cart_.T

fig = go.Figure(data=[
    go.Scatter3d(
        x=x, y=y, z=z, mode='markers',
        marker=dict(size=4, opacity=0.8)
    ),
    go.Scatter3d(
        x=xx, y=yy, z=zz, mode='markers',
        marker=dict(size=4, opacity=0.8)
    )
])
fig.update_layout(
    autosize=False,
    width=350,
    height=350,
    margin=dict(l=0, r=0, b=0, t=0)
)
fig.show()
```
