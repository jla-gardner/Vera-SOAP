```python
%config InlineBackend.figure_formats = ['svg']
import numpy as np
from soap_package5 import phi, expand_in_rbfs
import matplotlib.pyplot as plt
```

```python
rcut = 3.7
nmax = 4

xs = np.arange(0, rcut, 0.02)
ns = np.arange(nmax) + 1

phis = np.array([
    [phi(rcut, x, n) for x in xs]
    for n in ns
])

cmap = plt.get_cmap("viridis")
colours = [cmap(n/nmax) for n in ns]
for _phi, colour, n in zip(phis, colours, ns):
    plt.plot(xs, _phi, c=colour, label=n)
plt.legend(title="n=", loc="best")
plt.xlim(0, rcut)
plt.ylim(bottom=0)
plt.title("Un-normalised Radial Basis Functions");
```

```python
normalised_rbfs = np.array([
    expand_in_rbfs(rcut, nmax, x) for x in xs
]).T

for func, colour, n in zip(normalised_rbfs, colours, ns):
    plt.plot(xs, func, c=colour, label=n)
plt.hlines([0], 0, rcut, "k", linewidth=0.5)
plt.legend(title="n=", loc="best")
plt.ylim(-1, 2)
plt.xlim(0, rcut)
plt.title("Normalised Radial Basis Functions");
```
